### Dataset Shape

```python
df = df.drop_duplicates()
df.fillna(df.median(numeric_only=True), inplace=True)
```

## Data Cleaning

This step cleans the dataset by removing duplicates and handling missing values.


```python
df['G3'].mean()
```

### Average Final Grade

This calculates the average final math score of students.


```python
df[df['G3'] > 15].shape[0]
```

### Students Scoring Above 15

This counts how many students scored more than 15 in the final grade.


```python
df['studytime'].corr(df['G3'])
```

### Correlation Analysis

This checks the relationship between study time and final grade.


```python
df.groupby('sex')['G3'].mean()
```

### Gender-wise Performance

This compares average final grades between male and female students.


```python
plt.hist(df['G3'], bins=10)
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')
plt.title('Distribution of Final Grades')
plt.show()
```

### Histogram of Final Grades

This shows the distribution of students’ final grades.


```python
plt.scatter(df['studytime'], df['G3'])
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.title('Study Time vs Final Grade')
plt.show()
```

### Study Time vs Final Grade

This visualizes the relationship between study time and final grades.


```python
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade')
plt.title('Average Final Grade by Gender')
plt.show()
```
